```python
Q1. What is Flask Framework? What are the advantages of Flask Framework?
-->
Flask is a lightweight, open-source web framework written in Python that allows developers to build web applications quickly and easily. 
It provides developers with tools and libraries to create web applications that can be deployed on any web server that supports Python.

Advantages of Flask Framework:

a.Lightweight: 
    Flask is a micro web framework, which means it has a small codebase and requires fewer resources, making it easy to learn, use and maintain.
    
b.Flexibility: 
    Flask offers a lot of flexibility to developers, allowing them to add only the features they need,
    without having to include unnecessary components.
    
c.Easy to use: 
    Flask provides a simple and intuitive API, making it easy for developers to build web applications quickly and with fewer lines of code.
    
d.Modular design:
    Flask follows a modular design, allowing developers to use and extend only the parts they need, 
    making it easy to customize and scale their applications as needed.
    
e.Large community: 
    Flask has a large and active community of developers, making it easy to find support, documentation, and additional resources online.
```

```python
Q3. What is App routing in Flask? Why do we use app routes?
-->
    App routing is the process of defining the URL routes that map to specific functions in the application. 
    It allows you to specify what function should be executed when a particular URL is accessed by a client.
    
We use app routes in Flask for several reasons:
    
a.To organize our code: 
    By defining routes, we can separate the different parts of our application and make it easier to manage and maintainble.
    
b.To handle client requests:
    App routes are essential for handling incoming requests from clients and directing them to the appropriate functions.
    
c.To create user-friendly URLs:
    By defining routes, we can create user-friendly URLs that are easy to read and remember, rather than having long and complex URLs.
    
d.To support different HTTP methods: 
    Flask routes can be configured to handle different HTTP methods (such as GET, POST, PUT, DELETE), 
    allowing us to build more complex web applications that interact with a server in various ways.
```

```python
Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

-->
 url_for() :
        It generates a URL for the specified function and passes the arguments provided in the function call. 
        This function takes the name of the function as the first argument and any additional keyword arguments representing the URL variables.
        
from flask import Flask

app = Flask(__name__)

#Define a route with a parameter
@app.route('/user/<username>')
def show_user_profile(username):
    return 'User: {}'.format(username)

# Define a route that uses the url_for function
@app.route('/profile')
def profile():
    # Generate a URL for the show_user_profile function with the username argument
    user_url = url_for('show_user_profile', username='JohnDoe')
    return 'Profile page<br><a href="{}">{}</a>'.format(user_url, user_url)

if __name__=="__main__":
    app.run(host="0.0.0.0")
    
    
In this example, 

we define two routes - show_user_profile() and profile(). 

1.The show_user_profile() route takes a parameter username and returns the name of the user.
The profile() route uses the url_for() function to generate a URL for the show_user_profile() route and passes a value for the username parameter.

When we run this Flask application and navigate to http://https://salmon-waiter-lzqeg.pwskills.app:5000/profile,
we will see a link to the show_user_profile() route with the username parameter set to "JohnDoe".
When we click on this link, we will see the message "User: JohnDoe" displayed on the page. 
The url_for() function generates the correct URL with the appropriate parameter value based on the function name and the keyword arguments provided.
```

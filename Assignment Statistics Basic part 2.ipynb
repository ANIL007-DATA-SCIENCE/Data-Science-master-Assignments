```python
Q1. What are the three measures of central tendency?
-->
    The three measures of central tendency are : mean, median, and mode.
    
    
    
Q2. What is the difference between the mean, median, and mode? How are they used to measure the
central tendency of a dataset?
--> 
    Mean is the average value of a dataset and is calculated by adding up all the values in the dataset and dividing by the number of values. 
    Median is the middle value in a dataset when it is arranged in order. 
    Mode is the value that appears most frequently in a dataset. 
        
These measures are used to describe the typical or central value in a dataset.


    
Q3. Measure the three measures of central tendency for the given height data:
[178,177,176,177,178.2,178,175,179,180,175,178.9,176.2,177,172.5,178,176.5]
-->
    Mean = 176.86875
    Median = 177
    Mode = 178
    
    

Q4. Find the standard deviation for the given data:
[178,177,176,177,178.2,178,175,179,180,175,178.9,176.2,177,172.5,178,176.5]
-->
    Standard deviation = 2.2978
    
    

Q5. How are measures of dispersion such as range, variance, and standard deviation used to describe
the spread of a dataset? Provide an example.
-->
    Measures of dispersion such as range, variance, and standard deviation are used to describe the spread of a dataset:
        
        1.Range is the difference between the largest and smallest values in a dataset. 
        2.Variance measures how spread out the values in a dataset are from the mean, and standard deviation is the square root of the variance. 
        
    These measures can provide information about the variability of a dataset, which is important when comparing and analyzing data. 
    
    For example:
        if the standard deviation is large, it indicates that the data is spread out over a wide range of values.
        


Q6. What is a Venn diagram?
-->
    A Venn diagram is a graphical representation of set relationships. 
    It is a diagram that shows all possible logical relations between a finite collection of different sets.


    
Q7. For the two given sets A = (2,3,4,5,6,7) & B = (0,2,6,8,10). 
Find:
(i) A n B
(ii) A ⋃ B
-->
     (i) A B = {2,6}
     (ii) A ⋃ B = {0,2,3,4,5,6,7,8,10}
    


Q8. What do you understand about skewness in data?
-->
    Skewness in data refers to the degree of asymmetry of the distribution of values. 
    It is a measure of the deviation of a dataset from a normal distribution.
    

Q9. If a data is right skewed then what will be the position of median with respect to mean?
-->
    If a data is right skewed, the median will be to the left of the mean.
    
    
Q10. Explain the difference between covariance and correlation. How are these measures used in
statistical analysis?
-->
    Covariance measures the direction of the linear relationship between two variables, 
    while correlation measures both the strength and direction of the linear relationship between two variables.
    
    Both measures are used in statistical analysis to study the relationship between two variables.

Q11. What is the formula for calculating the sample mean? Provide an example calculation for a
dataset.
-->
    x̄ = (x1 + x2 + ... + xn) / n

    For example, if the dataset is {2, 4, 6, 8}, 
    the sample mean is:

    x̄ = (2 + 4 + 6 + 8) / 4 = 5


Q12. For a normal distribution data what is the relationship between its measure of central tendency?
--> 
    For a normal distribution, the mean, median, and mode are all equal.
    
    

Q13. How is covariance different from correlation?
-->
    Covariance is a measure of the direction of the linear relationship between two variables, 
    while correlation measures both the strength and direction of the linear relationship between two variables. 
    Covariance is affected by the scale of the variables, while correlation is not.


Q14. How do outliers affect measures of central tendency and dispersion? Provide an example.
-->
    Outliers can affect measures of central tendency and dispersion:
        1.If there are extreme values in a dataset, the mean can be pulled in the direction of the outlier, making it an unreliable measure of central tendency. 
        Similarly, outliers can affect measures of dispersion such as variance and standard deviation, which are sensitive to extreme values.
        For example, if a dataset has an outlier value of 100, the standard deviation will be larger than if the outlier value was removed.
```

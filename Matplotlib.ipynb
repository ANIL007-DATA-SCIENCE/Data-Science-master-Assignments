```python
Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
--> 
    Scatter plots are used to observe relationship between variables and uses dots to represent the relationship between them.
```

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
# Note: Also add title, xlabel, and ylabel to the plot.
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,c = '#2919E4', alpha=0.8)
plt.xlabel("This is my x axis")
plt.ylabel("This is my y axis")
plt.title(" x vs y")
```

```python
Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
-->
    bar plot is  bounded with a rectangle depending on the given parameters.
```

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,5))
plt.bar(company,profit)
plt.xlabel("company")
plt.ylabel("Profit in Cr")
plt.title(" Profit Per Year ")
```

```python
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
-->Box plot is helps to create box plots

A Box Plot is also known as Whisker plot is created to display the summary of the set of data values having properties like minimum, 
first quartile, median, third quartile and maximum. In the box plot, a box is created from the first quartile to the third quartile, 
a vertical line is also there which goes through the box at the median. 
Here x-axis denotes the data to be plotted while the y-axis shows the frequency distribution.
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data =[box1,box2]
fig = plt.figure(figsize =(8, 5))

plt.boxplot(data)
plt.show()
```

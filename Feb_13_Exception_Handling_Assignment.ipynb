Q1) Explain why we have to use the Exception class while creating a Custom Exception.
Note: Here Exception class refers to the base class for all the exceptions.

--> 
1) Standardization: 
    By inheriting from the Exception class, our custom exception will follow the standard exception hierarchy in Python. This makes it easier for other developers to understand and work with our code.

2) Consistency:
    By inheriting from the Exception class, our custom exception will behave like other built-in exceptions. It will have the same attributes and methods that are commonly used to handle exceptions in Python.

3) Customization: 
    The Exception class provides a set of attributes and methods that can be overridden in our custom exception. This allows us to customize the behavior of our exception to meet our specific needs.

4) Clarity: 
    By using the Exception class as the base class for our custom exception, we make it clear that our class is intended to be used for error handling. This makes our code more readable and easier to maintain.

Q4) Why LookupError class is used? Explain with an example KeyError and IndexError.

-->
LookupError: This exception is raised when a key or index is not found in a collection or mapping. 

KeyError: This exception is raised when attempting to access a dictionary key that does not exist

IndexError: This exception is raised when attempting to access a list or tuple index that is out of range.

```python
# Example of handling keyError
try:
    d = {"a": 1, "b": 2, "c": 3}
    value = d["d"]
except KeyError as e:
    print("This is a KeyError")
```

```python
# Example of handling IndexError
try: 
    my_list = [1, 2, 3]
    value = my_list[3]
except IndexError as e:
    print("This is IndexError")
```

```python
# LookupError handling Example
my_dict = {"a": 1, "b": 2, "c": 3}

try:
    value = my_dict["d"]
except LookupError as e:
    print("This is the LookupError ")
```

Q5) Explain ImportError. What is ModuleNotFoundError?

-->
ImportError is raised when a module or package cannot be imported.

This can happen for a variety of reasons, 
1.such as the module being misspelled, 
2.not installed, 
3.not available on the current system.

```python
# Example of ImportError
try:
    import non_existent_module
except ImportError as e:
    print("This is the ImportError")
```

ModuleNotFoundError: 

1.It is  a subclass of ImportError that was added in Python 3.6. 
2.It is raised when a module or package cannot be found during the import process.
3.It is a more specific version of ImportError that includes additional information about the missing module or package.

```python
# Example of ModuleNotFoundError
try:
    import non_existent_module
except ModuleNotFoundError as e:
    print("This is a ModuleNotFoundError ")
```
